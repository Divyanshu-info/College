```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special as sp
n= 3
d = 100
R = 3000
prr = 0 +10 * n * np.log10(d/R)
values = np.linspace(0.5, 0.9, 10)
sigma2 = np.linspace(0, 10, 30)[1:]

def va(prob):
    a = sp.erfinv(1 - 2*prob)
    return a

def vb(sigma):
    b = 10*n*np.log10(np.e)/sigma*np.sqrt(2)
    return b


for value in values:
    u_gamma =[]
    for sigmai in sigma2: 
        a = va(value)
        b= vb(sigmai)
        u_gamma.append(0.5*(1 - sp.erf(a) + np.exp((1 - 2*a*b)/b**2)*(1 - sp.erf((1 - a*b)/b))))
    plt.plot(u_gamma)
plt.xticks(ticks=np.linspace(0, 30, 6), labels=np.linspace(0,10, 6))
plt.legend(np.round(values, 2))
fig = plt.gcf()
fig.set_size_inches(16, 9)
plt.show()

```
