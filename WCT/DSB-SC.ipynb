```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
from scipy.fft import fft, fftfreq

N=1000
t = np.linspace(0,1,1000)
fm = 5
fc = 50
m = 1/2*np.sin(2*np.pi*fm*t)
c = np.cos(2*np.pi*fc*t)

y=m*c

s1 = y*c
b, a = signal.butter(5, 0.1)
s2 = signal.lfilter(b,a,s1)

plt.subplot(5, 1, 1)
plt.title('DSB-SC Modulation')
plt.plot(m, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')


plt.subplot(5, 1, 2)
plt.plot(c, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')


plt.subplot(5, 1, 3)
plt.plot(y, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('DSB-SC Modulated signal')

plt.subplot(5, 1, 4)
plt.plot(s2, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('DSB-SC Demodulated signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(32, 20)

plt.show()
# fig.savefig('Amplitude Modulation.png', dpi=100)

yf1 = fft(y)
xf1 = fftfreq(N, 1/1000)[:N//2]

yf2 = fft(s2)
xf2 = fftfreq(N, 1/1000)[:N//2]

plt.subplot(2, 1, 1)
plt.plot(xf1, 2.0/N * np.abs(yf1[0:N//2]), color="teal")
plt.xlabel('DSB-SC Modulated signal Frequency Representation')

plt.subplot(2, 1, 2)
plt.plot(xf2, 2.0/N * np.abs(yf2[0:N//2]), color="indigo")
plt.xlabel('DSB-SC Demodulated signal Frequency Representation')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(32, 10)

plt.show()
```
