```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal


N = 1000
t = np.linspace(0, 1, 1000)
fm = 5
fc = 50
Am = 5
Ac = 5
b= 5
b2 = 5

message = Am*np.cos(2*np.pi*fm*t)
carrier = Ac*np.cos(2*np.pi*fc*t)


freq_modulated =Ac*np.cos(2*np.pi*fc*t + b*Am*np.sin(2*np.pi*fm*t)/fm)

phase_modulated = Ac*np.cos(2*np.pi*fc*t + b2*np.cos(2*np.pi*fm*t))

wave = freq_modulated = Ac*np.cos(2*np.pi*fc*t + b*np.sin(2*np.pi*fm*t))

mt = np.cos(2*np.pi*fm*t)
integrated_msg_wave = np.cumsum(mt)
fmi = np.cos(2*np.pi*fc*t + b*integrated_msg_wave)

plt.subplot(5, 1, 1)
plt.title('FM Modulation')
plt.plot(message, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')


plt.subplot(5, 1, 2)
plt.plot(carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')


plt.subplot(5, 1, 3)

plt.plot(freq_modulated, color="purple")
plt.plot(message, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Frequency Modulated signal')

plt.subplot(5, 1, 4)
plt.plot(phase_modulated, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('Phase Demodulated signal')

plt.subplot(5, 1, 5)
plt.plot(fmi, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('FM wave from PM wave signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(32, 20)

plt.show()
```
