```python
from scipy import stats
import numpy as np

mu=0
sigma=1 
L=100000 

samples_normal = np.random.normal(loc=mu,scale=sigma,size=(L,1)) 

import matplotlib.pyplot as plt

fig, ax0 = plt.subplots(ncols=1, nrows=1) 
(values, bins, _) = ax0.hist(samples_normal,bins=100,density=True,label="Histogram of samples") 

bin_centers = 0.5*(bins[1:] + bins[:-1])
pdf = stats.norm.pdf(x=bin_centers, loc=mu, scale=sigma)
ax0.plot(bin_centers, pdf, label="PDF", color='black') 
ax0.legend()  
ax0.set_title('PDF of samples');

plt.grid()
plt.rc('font', size=15)
plt.subplots_adjust(hspace=1)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```
