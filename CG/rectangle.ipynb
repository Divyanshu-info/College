```python
import numpy as np
import cv2
shape = (1000, 1000, 3)


def draw_coordinates():
    graph = np.zeros(shape, dtype="uint8")
    for i in range(10):
        for j in range(10):
            graph = cv2.add(graph, cv2.circle(graph, (i*100, j*100),
                                              radius=1, color=(255, 255, 255), thickness=-1))
    return graph


canvas = np.zeros(shape, dtype="uint8")

blue = (255, 0, 0)
cv2.rectangle(canvas, (300, 100), (500, 200), blue, -1)
rotate_matrix = cv2.getRotationMatrix2D(center=(500, 500), angle=45, scale=1)

image = cv2.warpAffine(src=canvas, M=rotate_matrix, dsize=(1000, 1000))

cv2.putText(image, '(5, 5)', (500, 520),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
cv2.imwrite("Rotation1.png", draw_coordinates() + canvas + image)
while(True):
    cv2.imshow("Rotation1", mat=draw_coordinates() + canvas + image)

    if cv2.waitKey(0) == 27:
        cv2.destroyAllWindows()
        break
```

```python
import numpy as np
import cv2
shape = (1600, 1600, 3)


def draw_coordinates():
    graph = np.zeros(shape, dtype="uint8")
    for i in range(int(shape[0]/100)):
        for j in range(16):
            graph = cv2.add(graph, cv2.circle(graph, (i*100, j*100),
                                              radius=1, color=(255, 255, 255), thickness=-1))
    graph = cv2.add(graph, cv2.circle(graph, (400, 150),
                                      radius=1, color=(255, 255, 255), thickness=-1))
    return graph


canvas = np.zeros(shape, dtype="uint8")

blue = (255, 0, 0)
cv2.rectangle(canvas, (300, 100), (500, 200), blue, -1)
t3x3 = [[1, 0, 0], [0, 1, 0], [-5, -5, 1]]
sl3x3 = [[3, 0, 0], [0, 1, 0], [0, 0, 1]]
t2_3x3 = [[1, 0, 0], [0, 1, 0], [5, 5, 1]]
b_matrix = [[3, 5, 5, 3], [1, 1, 2, 2], [0, 0, 0, 0]]


f_matrix = np.matmul(np.matmul(np.matmul(t3x3, sl3x3), t2_3x3), b_matrix)
f_matrix = f_matrix[0:2][:]
f_matrix = f_matrix.transpose()
pts = np.array([f_matrix], np.int32)
pts *= 100
cv2.add(canvas, cv2.fillPoly(canvas, pts, blue))
cv2.imwrite("Scaling.png", draw_coordinates() + canvas)
while(True):
    cv2.imshow("Scaling", mat=draw_coordinates() + canvas)

    if cv2.waitKey(0) == 27:
        cv2.destroyAllWindows()
        break
```
